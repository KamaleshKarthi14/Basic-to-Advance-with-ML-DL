```python
type_label
```

```python
#Fit Model
```

```python
model=svm.SVC(kernel='linear')
```

```python
model.fit(ingredients,type_label) #train
```

```python
w= model.coef_[0] 
a=-w[0]/w[1]
xx=np.linspace(30,60)
yy=a*xx-(model.intercept_[0]/w[1])
```

```python
sns.lmplot('Flour','Sugar',data=recipes,hue='Type',palette='Set1',fit_reg=False,scatter_kws={"s":70})
plt.plot(xx,yy,linewidth=2,color='black')
```

```python
#Support Vetor 
b=model.support_vectors_[0]
yy_down = a * xx +(b[1]-a*b[0])
b=model.support_vectors_[-1]
yy_up = a * xx +(b[1]-a*b[0])
```

```python
sns.lmplot('Flour','Sugar',data=recipes,hue='Type',palette='Set1',fit_reg=False,scatter_kws={"s":70})
plt.plot(xx,yy,linewidth=2,color='black')
plt.plot(xx,yy_down,'k--')
plt.plot(xx,yy_up,'k--')
```

### Elbow Method

```python
x=dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11): #10
  model=KMeans(n_clusters=i,init='k-means++',random_state=0)
  model.fit(x)
  wcss.append(model.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel("NO. of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
model2=KMeans(n_clusters=5,init='k-means++',random_state=0)
```

```python
label=model2.fit_predict(x)
```

```python
label
```

```python
plt.scatter(x[label==0,0],x[label==0,1],s=100,c='red',label='Cluster1')
plt.scatter(x[label==1,0],x[label==1,1],s=100,c='blue',label='Cluster2')
plt.scatter(x[label==2,0],x[label==2,1],s=100,c='green',label='Cluster3')
plt.scatter(x[label==3,0],x[label==3,1],s=100,c='cyan',label='Cluster4')
plt.scatter(x[label==4,0],x[label==4,1],s=100,c='magenta',label='Cluster5')
plt.scatter(model2.cluster_centers_[:,0],model2.cluster_centers_[:,1],s=300,c='black',label='centroid')
#plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel("NO. of Clusters")
plt.ylabel("WCSS")
plt.legend()
plt.show()
```
